**Yearly sales trend**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Year'] = df['Order Date'].dt.to_period('Y')
df['Year']
```

```python
sns.barplot(data=df, x='Year', y= 'Sales Amount')
plt.show()
```

**Monthly Sales Trend**

```python
df['Month'] =df['Order Date'].dt.to_period('M')
```

```python
monthly_sales=df.groupby('Month')['Sales Amount'].sum().reset_index().astype(str)
monthly_sales
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Sales Amount', marker='o', color='b', linewidth=2.5)
plt.title('Monthly Sales Performance Trends')
plt.xlabel('Timeline (Year-Month)')
plt.ylabel('Total Sales Volume ($)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Business insights and recommendations*

Business Insights
1. Products have more sales in North, south , and in west regions.
2. Aggressive discounting increases sales volume but significantly erodes profit margins.
3. Investing in inventory and marketing campaigns for Electronics can significantly boost overall revenue.
4. Sales campaigns during festive and holiday seasons can maximize revenue

Recommedations
1. Limit discounts above 20%.
2. Launch region-specific marketing campaigns.
3. Increase stock levels for Electronics and other top-selling products
4. Promote Second Class Shipping as the default delivery option & free shipping to encourage larger orders.
